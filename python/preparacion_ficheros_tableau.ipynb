```python
####Preparación ficheros para cargar en Tableau. Los ficheros de coordenadas para RFM y para la predicción tienen
####un campo con las dos coordenadas separadas por comas. En este fichero crearemos dos columnas nuevas, una con
####la latitud y otra con la longitud.

##Importamos libreria pandas
import pandas as pd

##Cargamos fichero de las coordenadas RFM
terremotos_coordenadas_RFM_tableau = pd.read_csv("../files/terremotos_coordenadas_RFM.csv", sep=',')
##Eliminamos la primera columna del DF
terremotos_coordenadas_RFM_tableau.pop('Unnamed: 0')
terremotos_coordenadas_RFM_tableau.head()
```

```python
##Obtener latitud
def separa_coordenadas_latitud(cadena):
    lista = cadena.split(',')
    return lista[0]

##Obtener longitud
def separa_coordenadas_longitud(cadena):
    lista = cadena.split(',')
    return lista[1]

##Creamos las columnas de latitud y longitud
terremotos_coordenadas_RFM_tableau['latitud'] = terremotos_coordenadas_RFM_tableau['coordenadas'].apply(separa_coordenadas_latitud)
terremotos_coordenadas_RFM_tableau['longitud'] = terremotos_coordenadas_RFM_tableau['coordenadas'].apply(separa_coordenadas_longitud)
```

```python
##Guardamos en un nuevo fichero la segmentación RFM con latitud y longitud
terremotos_coordenadas_RFM_tableau.to_csv("../tableau/terremotos_latitud_longitud_RFM_tableau.csv")
terremotos_coordenadas_RFM_tableau.head()
```

```python
###Creamos el fichero de predicciones incluyendo las columnas de latitud y longitud
##Cargamos el fichero de prediccion de coordenadas
terremotos_coordenadas_prediccion_tableau = pd.read_csv("../files/prediccion_terremotos_coordenadas.csv", sep=',')
##Eliminamos la primera columna
terremotos_coordenadas_prediccion_tableau.pop('Unnamed: 0')
##Creamos columnas de latitud y longitud
terremotos_coordenadas_prediccion_tableau['latitud'] = terremotos_coordenadas_prediccion_tableau['coordenadas'].apply(separa_coordenadas_latitud)
terremotos_coordenadas_prediccion_tableau['longitud'] = terremotos_coordenadas_prediccion_tableau['coordenadas'].apply(separa_coordenadas_longitud)
##Guardamos el nuevo DF con latitud y longitud en un nuevo fichero
terremotos_coordenadas_prediccion_tableau.to_csv("../tableau/terremotos_latitud_longitud_prediccion_tableau.csv")
terremotos_coordenadas_prediccion_tableau.head()
```
