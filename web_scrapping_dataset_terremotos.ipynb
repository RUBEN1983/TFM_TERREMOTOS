```python
###Carga de ficheros
import pandas as pd

fechas=pd.read_csv("files/dataset_fechas.csv", sep=";")
```

```python
###Cargamos beautiful soup
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
###comandos para extraer los datos de la web, generamos ficheros con los datos de los terremotos
subset = fechas[['Fecha_inicio', 'Fecha_fin']]
tuples = [tuple(x) for x in subset.values]
for x,y in tuples:
    texto = requests.get('http://www.seismicportal.eu/fdsnws/event/1/query?start=%s&end=%s' %(x,y)).content
    archi= open('%s_%s.xml' %(x,y),'w')
    archi.write(texto)
    archi.close()
```
