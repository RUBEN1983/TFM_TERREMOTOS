```python
##Creación Dataframe coordenadas para RFM con los valores a cero 

###Función auxiliar para eliminar algunos caracteres y espacios
def elimina_caracteres(cadena):
    
    cadena=cadena.strip('(')
    cadena=cadena.strip(')')
    cadena=cadena.replace("'","")
    cadena=cadena.replace(" ","")
    
    return cadena

###Función para la creación de DataFrame de coordenadas, Recencia, Frecuencia y Magnitud
def dataframe_coordenadas_RFM():
    
    ##Definimos longitudes negativas (-179 a -1) y positivas (0 a 179)
    lista_longitud_negativa = list(range(-179,0,1))
    lista_longitud_positiva = list(range(0,180,1))
    
    ##Definimos latitudes negativas (-89 a -1) y positivas (0 a 89)
    lista_latitud_negativa = list(range(-89,0,1))
    lista_latitud_positiva = list(range(0,90,1))
    
    ##Tendremos un valor de longitud y latitud que valga -0
    cero_negativo ='-0'
    
    ##Lista de coordenadas vacía
    coordenadas = []
    
    ##Bucle para recorer las latitudes negativas (-89 a -1)
    for i in lista_latitud_negativa:
        ##Bucle para recorrer las longitudes negativas (-179 a -1)
        for j in lista_longitud_negativa:
            coordenadas.append(str((i,j)))
        ##Creamos longitud -0
        coordenadas.append(str((i,"-0")))
        ##Bucle para recorrer las longitudes positivas (0 a 179)
        for k in lista_longitud_positiva:
            coordenadas.append(str((i,k)))
            
    ##Bucle para crear latitud -0 con longitudes negativas (-179 a -1)      
    for i in lista_longitud_negativa:
        coordenadas.append(str((cero_negativo,i)))
    ##Creamos latitud, longitud (-0,-0)
    coordenadas.append(str((cero_negativo,cero_negativo)))
    ##Bucle para crear latitud -0 con longitudes positivas (0 a 179)
    for i in lista_longitud_positiva:
        coordenadas.append(str((cero_negativo,i)))
        
    ##Bucle para recorer las latitudes positivas (0 a 89)   
    for i in lista_latitud_positiva:
        ##Bucle para recorrer las longitudes negativas (-179 a -1)
        for j in lista_longitud_negativa:
            coordenadas.append(str((i,j)))
        ##Creamos longitud -0
        coordenadas.append(str((i,cero_negativo)))
        ##Bucle para recorrer las longitudes positivas (0 a 179)
        for k in lista_longitud_positiva:
            coordenadas.append(str((i,k)))
    
    ##Creamos columna de coordenadas y aplicamos la función para eliminar caracteres
    dataframe_coordenadas = pd.DataFrame(coordenadas, columns=["Coordenadas"])
    dataframe_coordenadas["Coordenadas"]=dataframe_coordenadas["Coordenadas"].apply(elimina_caracteres)
    
    ##Creamos columnas Recencia, Frecuencia y Magnitud
    dataframe_coordenadas['RECENCIA'] = 0
    dataframe_coordenadas['FRECUENCIA'] = 0
    dataframe_coordenadas['MAGNITUD'] = 0
    
    ##Retornamos DataFrame coordenadas
    return dataframe_coordenadas
```

```python
###Retornamos el DF creado en el fichero "coordenadas_RFM_ceros.csv"
df_coordenadas_RFM = dataframe_coordenadas_RFM()
df_coordenadas_RFM.to_csv('../files/coordenadas_RFM_ceros.csv',sep=';')
df_coordenadas_RFM.head()
```
