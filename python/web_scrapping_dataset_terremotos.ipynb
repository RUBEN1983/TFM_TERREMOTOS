```python
###Importamos libreria pandas
import pandas as pd

###Cargamos en un DF el csv con las fechas comprendidas entre 01/01/1998 y 31/12/2016
fechas=pd.read_csv("../files/dataset_fechas.csv", sep=";")
fechas.head()
```

```python
###Importamos librerias requests, BeautifulSoup y lxml para hacer web scrapping
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
###Comandos para hacer web scrapping
subset = fechas[['Fecha_inicio', 'Fecha_fin']]
#Convertimos cada fila del DF de fechas en una tupla con fechas inicio y fin de cada fichero
tuples = [tuple(x) for x in subset.values]
#Bucle for para realizar todas las llamadas al WebService y generar los ficheros XML
for x,y in tuples:
    texto = requests.get('http://www.seismicportal.eu/fdsnws/event/1/query?start=%s&end=%s' %(x,y)).content
    archi= open('%s_%s.xml' %(x,y),'w')
    archi.write(texto)
    archi.close()
```
